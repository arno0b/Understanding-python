````python
def func1(**d):
    list1 = []
    for i in d.values():
        if type(i) == dict: 
            for j in i.values():
                list1.append(j)
        elif type(i) == list: 
            for j in i:
                list1.append(j)
        elif type(i) == tuple: 
            for j in i:
                list1.append(j)                
        else:
            list1.append(i)
    
    return list1


````
