The difference is due to the double square brackets `([[]])` used in `df1[['Year']]`. The double square brackets indicate that we want to select one or more columns as a DataFrame, while a single set of square brackets `([])` indicates that we want to select a single column as a Series.

For example, if df1 is a DataFrame with columns '`Year`', '`Month`', and '`Day`', then `df1['Year']` will return a Series with the values in the '`Year`' column, while ``df1[['Year']]` will return a DataFrame with a single column '`Year`'.

```python
df2 = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2015_totals.html')
df2
```

```python
df2[0]
```

`df2[0]` accesses the first element of the list `df2`. The first element of `df2` is a Pandas dataframe, which is why it can be displayed as a dataframe. The `pd.read_html` function returns a list of dataframes, one for each table found in the HTML content at the given URL. In this case, there is only one table on the page, which is why the result is stored as `df2[0]`. If there were multiple tables on the page, you could access the other tables by indexing the list with their respective indices, for example `df2[1]`, `df2[2]`, etc.

dictionary to dataframe

```python
d1 = {
      'key1' : [1,2,3,4,5,6],
      'key2' : [7,8,9,10,11,12],
      'key3' : [13,14,15,16,17,18],
      'key4' : [19,20,21,22,23,24]
      }
```

```python
df1 = pd.DataFrame(d1, index=['a','b','c','d','e','f'])
df1
```

keys are column and values are rows. all value array length must be same. But the array can be of different dtypes



```python
d2 = {
      'key1' : [1,2,3,4,5,'arnob'],
      'key2' : [7,8,'sakif',10,11,12],
      'key3' : [13,14,15,16,17,18],
      'key4' : [19,20,21,'json','positron',24]
      }
```

```python
df2 = pd.DataFrame(d2)
df2
```

only common rows are shown

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how = 'left', on = 'key4')
```

The resulting DataFrame would have the following columns and values:

- '`key1_x`': values from the '`key1`' column in `df1`
- '`key2_x`': values from the '`key2`' column in `df1`
- '`key3_x`': values from the '`key3`' column in `df1`
- '`key4`': values from the '`key4`' column in `df1`
- '`key1_y`': values from the '`key1`' column in `df2` (filled with NaN values where no matching values were found)
- '`key2_y`': values from the '`key2`' column in `df2` (filled with NaN values where no matching values were found)
- '`key3_y`': values from the '`key3`' column in `df2` (filled with NaN values where no matching values were found)
