## Modules and packages ##

A module is a collection of functions and/or classes that are packaged together in a single file. It can be imported into other parts of a program to make use of its functionality.

A package is a collection of modules that are grouped together in a directory. It is a way to organize and structure code in a large project. Packages can contain multiple modules, and each module can contain multiple functions and/or classes.

A package is a higher-level construct compared to modules, instances and functions. It is used to organize the structure of the codebase and make it more maintainable.

The first line creates an instance of the logging module's logger class using the __name__ variable as the name of the logger.

The second line 
```python
logger.setLevel(logging.DEBUG)
``` 
sets the level of the logger to DEBUG, meaning that the logger will log all messages at the DEBUG level and above.

The third line 
```python
format = logging.Formatter('%(asctime)s %(levelname)s %(message)s')
``` 
creates a formatter object that will format the log messages in a specific way. In this case, the log messages will be formatted as "[asctime] [levelname] message".

The fourth line 
```python
file_handler = logging.FileHandler('employee.log')
``` 
creates a FileHandler object that will handle writing log messages to a file. The file that the FileHandler will write to is "employee.log".

The fifth line 
```python
file_handler.setFormatter(format)
``` 
sets the formatter of the file handler to the formatter object created in the third line, so the log messages will be written in the format specified by the formatter.

The last line 
```python
logger.addHandler(file_handler)
``` 
adds the file handler to the logger object, so the logger will use the file handler to write log messages to the "employee.log" file.

**Stream Handler**

Stream handler shows the log in the console instead of file for file handler

````python
import logging

#create custom logger
logger = logging.getLogger(__name__)

#create  handler
c_handler = logging.StreamHandler()
f_handler = logging.FileHandler('file_handler.log')
c_handler.setLevel(logging.WARNING)
f_handler.setLevel(logging.ERROR)

#Create formatter and add it to handler

c_format = logging.Formatter('%(levelname)s %(name)s %(message)s')
f_format = logging.Formatter('%(asctime)s %(levelname)s %(name)s %(message)s')
c_handler.setFormatter(c_format)
f_handler.setFormatter(f_format)

#Add handler to the logger

logger.addHandler(c_handler)
logger.addHandler(f_handler)

logger.warning('warned')
logger.error('error MAPE')

````

The above code imports the `logging` module, which is a built-in python library that allows you to log messages to various outputs.

It creates a custom logger named `logger` using the `logging.getLogger(__name__)` method. `__name__` is a built-in variable that contains the name of the current module.

It creates two handlers, `c_handler` and `f_handler`, which are used to specify where the log messages should be output. `c_handler` is a `StreamHandler` that logs messages to the console and `f_handler` is a `FileHandler` that logs messages to a file named 'file_handler.log'.

It then sets the logging level for each handler. `c_handler` is set to log only warning messages and above, and `f_handler` is set to log only error messages and above.

It creates two formatters, `c_format` and `f_format`, which are used to specify the format of the log messages. The log message format is specified using a string that contains placeholders for various log message attributes such as the level name, the logger name, and the message itself.

It then sets the formatter for each handler using the `setFormatter()` method.

Finally, it adds the handlers to the logger using the `addHandler()` method and logs a warning and an error message using the `logger.warning()`.

````python
import logging

logger = logging.getLogger(__name__)
handler = logging.StreamHandler()

formatter = logging.Formatter('%(asctime)s %(name)s %(levelname)s %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)
logger.setLevel(logging.INFO)


def divide(dividend, divisor):
    try:
        logger.info(f"Dividing {dividend} by {divisor}")
        return dividend / divisor
    except ZeroDivisionError:
        logger.info('Zero division error')

print(divide(8,0))
````
