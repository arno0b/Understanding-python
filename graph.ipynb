```python
import cufflinks as cf
import pandas as pd 
import numpy as np
import plotly
import seaborn as sns
#pd.options.plotting.backend = "plotly"
cf.go_offline()
```

```python
df =pd.DataFrame(np.random.randn(100,3), columns = ['A','B','C'])
df
```

```python
df.iplot()
```

```python
df[['A']].iplot()
```

```python
df.iplot(x = 'A', y = 'B', mode= 'markers')
```

```python
df.iplot(kind = 'bubble', x = 'A', y = 'B', size = 'C')
```

```python
df1 = sns.load_dataset('titanic')
df1
```

```python
df1.iplot( x = 'sex', y = 'survived', kind = 'bar')
```

```python
df1[['age','fare']].iplot(x = 'age', y = 'fare', mode = 'markers', size = 5)
```

```python
df2 = sns.load_dataset('tips')
df2
```

```python
df2[['total_bill','tip', 'smoker']].iplot(x = 'total_bill', y = 'tip', colors=['blue', 'red'], categories = 'smoker', mode = 'markers')
```

3D axis

```python
df2.iplot(kind = 'scatter3d', x= 'total_bill', y = 'tip', z= 'size')
```

```python
sns.relplot(data= df2, x = 'total_bill', y = 'tip', hue= 'smoker')
```

```python
sns.relplot(data= df2, x = 'total_bill', y = 'tip', hue= 'smoker', style= 'size')
```

```python
df2[['total_bill','tip', 'smoker']].iplot(x = 'total_bill', y = 'tip', colors=['blue', 'red'], categories = 'smoker', mode = 'markers')
```
